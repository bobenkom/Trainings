Подготовить данные для визуализации. Источник данных https://www.kaggle.com/datasets/START-UMD/gtd?resource=download

```python
import pandas as pd

df = pd.read_csv('globalterrorism.csv', encoding="ISO-8859-1")
```

```python
df
```

```python
data = df[["eventid", "iyear", "country_txt"]]

def change_country(x):
    if x == "East Germany (GDR)" or x == "West Germany (FRG)":
        return "Germany"
    return x

data["country_txt"] = data["country_txt"].apply(change_country)
```

```python
a = data.groupby(by=["iyear", "country_txt"]).count()
a
```

```python
b = data.groupby(by=["country_txt"]).count().sort_values(by="eventid", ascending=False).head(30)
top_30_countries = list(b.index)
```

```python
terrorism = data.loc[data['country_txt'].isin(top_30_countries)]
result = terrorism.groupby(by=["iyear", "country_txt"]).count()
result
```

```python
result.to_csv('terrorism_result.csv') 
```
